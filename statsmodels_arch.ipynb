## Serie de retornos


```python
import yfinance as yf
import pandas as pd

serie = yf.download('EURUSD=X', start='2015-01-01', end='2023-12-31')['Adj Close']
retornos = serie.pct_change().dropna()
retornos.describe()
```

## Modelo ARIMA con Statsmodels


```python
from statsmodels.tsa.arima.model import ARIMA

modelo_arima = ARIMA(retornos, order=(1, 0, 1)).fit()
modelo_arima.summary()
```

## Modelo GARCH(1,1) con Arch


```python
from arch import arch_model

modelo_garch = arch_model(retornos * 100, mean='Zero', vol='GARCH', p=1, q=1)
resultado_garch = modelo_garch.fit(disp='off')
resultado_garch.summary()
```

Las salidas resumen capturan la dinámica de medias y volatilidades condicionales de la serie.

## Descarga y preparación de datos


```python
import backtrader as bt
import yfinance as yf
import pandas as pd

data = yf.download('QQQ', start='2018-01-01', end='2023-12-31')
feed = bt.feeds.PandasData(dataname=data)
```

## Estrategia de cruce de medias


```python
class SmaCross(bt.Strategy):
    params = dict(fast=20, slow=50)

    def __init__(self):
        sma1 = bt.ind.SMA(period=self.p.fast)
        sma2 = bt.ind.SMA(period=self.p.slow)
        self.crossover = bt.ind.CrossOver(sma1, sma2)

    def next(self):
        if not self.position and self.crossover > 0:
            self.buy()
        elif self.position and self.crossover < 0:
            self.close()
```

## Ejecución del backtest


```python
cerebro = bt.Cerebro()
cerebro.adddata(feed)
cerebro.addstrategy(SmaCross)
cerebro.broker.setcash(100000)
resultado = cerebro.run()
portafolio_final = cerebro.broker.getvalue()
{'valor_final': portafolio_final}
```

Backtrader permite añadir múltiples estrategias, comisiones, slippage y realizar optimizaciones.

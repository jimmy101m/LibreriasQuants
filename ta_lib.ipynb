## Descarga de datos


````python
import yfinance as yf
import talib
import pandas as pd

precio = yf.download('AAPL', start='2020-01-01', end='2023-12-31')
precio.tail()
````

## Indicadores técnicos


````python
precio['RSI_14'] = talib.RSI(precio['Adj Close'], timeperiod=14)
precio['SMA_50'] = talib.SMA(precio['Adj Close'], timeperiod=50)
precio['BB_upper'], precio['BB_middle'], precio['BB_lower'] = talib.BBANDS(precio['Adj Close'], timeperiod=20, nbdevup=2, nbdevdn=2)
precio[['Adj Close', 'RSI_14', 'SMA_50', 'BB_upper', 'BB_lower']].tail()
````

Con estos indicadores puedes prototipar reglas de trading o alimentar modelos cuantitativos.
